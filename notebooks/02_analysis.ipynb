```python
import sys
from pathlib import Path

PROJECT_ROOT = Path("..").resolve()
sys.path.append(str(PROJECT_ROOT))

from src.graph_builder import load_graph
from src.metrics import (
    compute_basic_metrics,
    compute_small_world_sigma,
    fit_power_law,
    compute_centrality,
    detect_communities,
    save_metrics_json,
    save_centrality_csv,
)

OUTPUT_DIR = PROJECT_ROOT / "output"
(OUTPUT_DIR / "tables").mkdir(parents=True, exist_ok=True)

COUNTRY = "GB"  # đổi thành "GB" để chạy lại cho dataset Anh
```

```python
G = load_graph(COUNTRY, OUTPUT_DIR / "graphs")
print(f"Số node: {G.number_of_nodes():,}")
print(f"Số edge: {G.number_of_edges():,}")
```

```python
basic_metrics = compute_basic_metrics(G)
for k, v in basic_metrics.items():
    print(f"{k}: {v}")

save_metrics_json(basic_metrics, COUNTRY, OUTPUT_DIR / "tables", "basic_metrics.json")
```

```python
sigma_result = compute_small_world_sigma(G, n_random=5)
for k, v in sigma_result.items():
    print(f"{k}: {v}")
```

```python
powerlaw_result = fit_power_law(G)
for k, v in powerlaw_result.items():
    print(f"{k}: {v}")

save_metrics_json(powerlaw_result, COUNTRY, OUTPUT_DIR / "tables", "powerlaw.json")
```

```python
centrality_df = compute_centrality(G)
print("Top 10 channel theo PageRank:")
centrality_df.head(10)
```

```python
community_result = detect_communities(G)
print(f"Số cộng đồng: {community_result['num_communities']}")
print(f"Modularity: {community_result['modularity']:.4f}")

# Gắn community_id vào bảng centrality rồi lưu lại
centrality_df["community"] = centrality_df["channel"].map(community_result["partition"])
save_centrality_csv(centrality_df, COUNTRY, OUTPUT_DIR / "tables")
centrality_df.head(10)
```
